## Data Visualisation

### Sales Analysis 

1. Analyze number of products sold and revenue generated to understand sales performance over time.

2. Identify customer demographics to determine which groups are purchasing the most products.

3. Track availability and stock levels to ensure the right products are in stock when customers are ready to buy.

```python
df['Revenue_generated'] = df['Revenue_generated'].round(2)
product_sold = df.groupby(['Product_type'])[['Number_of_products_sold', 'Revenue_generated']].sum().reset_index()
product_sold['Revenue_generated'] = product_sold['Revenue_generated'].round(2)

print(product_sold)
```

```python
plt.figure(figsize = (8,4))
colors = ['lavender','brown','orange']
pie_chart = plt.pie(product_sold['Number_of_products_sold'], 
                    labels = product_sold['Product_type'],
                    autopct = '%.2f%%',
                    wedgeprops ={ 'linewidth' : 3.0, 'edgecolor': 'white'},
                    textprops = {'size' : 'x-large'},
                    shadow = True,
                    colors = colors)
plt.title('percent of product sold by product type', fontsize= 15)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(
    x=product_sold['Product_type'],
    height=product_sold["Number_of_products_sold"],
    color='darkred',
    label='Number of Products Sold'
)
plt.bar(
    x=product_sold['Product_type'],
    height=product_sold["Revenue_generated"],
    bottom=product_sold["Number_of_products_sold"],
    color='orange',
    label='Revenue Generated'
)
# Titles and labels
plt.title("Revenue Generated vs. Number of Products Sold by Product Type", fontsize=12)
plt.xlabel("Product Type")
plt.ylabel("Stacked Values (Units + Revenue)")
plt.legend()
plt.tight_layout()
plt.show()
```

So, the highest number of products sold of the three product categories is skincare, which means 45% of business comes from skincare, 29% from haircare, and 25% from cosmetics.
And most of the revenue comes from skincare products, followed by haircare, and then cosmetics products.

In contrast, cosmetics and haircare have higher stock levels and lower availability. This indicates that the company cannot quickly ship these products to customers because it takes time to manufacture them.

## operations analysis:

•Analyze lead times, order quantities, and production volumes to improve inventory management and cut down on stockouts.

•Keep track of manufacturing lead time and costs to find areas for improvement and save money.

•Watch inspection results and defect rates to spot quality issues and enhance manufacturing processes.

```python
df.columns
```

```python
product = df.groupby(['Product_type'])[['Lead_time', 'Order_quantities', 'Production_volumes']].mean().reset_index()
product['Order_quantities'] = product['Order_quantities'].round(2)
product['Lead_time'] = product['Lead_time'].round(2)
product['Production_volumes'] = product['Production_volumes'].round(2)
product
```

•Skincare products come in larger quantities and take more time to produce. They also have a higher production volume; this refers to the number of products made by the company. Higher production volumes can mean longer lead times to ensure there is enough time to make the products and satisfy customer demand. 

•Haircare products also have a longer lead time and greater production volumes. This may be due to the need for more specific ingredients or manufacturing methods.

```python
avg_costs = df.groupby(['Manufacturing_lead_time'])['Manufacturing_costs'].mean().reset_index().sort_values(by = 'Manufacturing_costs')
avg_costs['Manufacturing_costs'] = avg_costs['Manufacturing_costs']
avg_costs
```

```python
avg_costs['Manufacturing_costs'] = avg_costs['Manufacturing_costs'].round(2)
plt.figure(figsize=(20, 10))
p = sns.barplot(x='Manufacturing_lead_time', y='Manufacturing_costs', data=avg_costs, palette='cividis')
for container in p.containers:
    p.bar_label(container, padding=-40, color='white', fontsize=12)

plt.title('Manufacturing Lead Time vs Cost', fontsize=15)
plt.xlabel('Manufacturing Lead Time')
plt.ylabel('Average Manufacturing Cost')
plt.tight_layout()
plt.show()
```

### Shipping Analysis:

•Analyze costs, transportation methods, and routes to improve logistics and lower shipping costs.

•Monitor shipping times, carriers, and transportation methods to make sure products arrive on time for customers.

•Track shipping costs related to carriers and revenue earned to find opportunities for savings.

```python
shipping = df.groupby(['Shipping_carriers'])['Shipping_costs'].sum().reset_index()
shipping
```

```python
# Prepare data (example: group by carrier and sum shipping costs)
shipping = df.groupby('Shipping_carriers')[['Shipping_costs']].sum().reset_index()

# Define valid hex colors
colors = ['#40E0D0', '#9FE2BF', '#CCCCFF']  # turquoise, seafoam green, lavender blue

# Plot pie chart
plt.figure(figsize=(6, 4))
plt.pie(
    shipping['Shipping_costs'],
    labels=shipping['Shipping_carriers'],
    autopct='%.2f%%',
    wedgeprops={'linewidth': 2.0, 'edgecolor': 'white'},
    textprops={'size': 'x-large'},
    shadow=True,
    colors=colors[:len(shipping)]  # ensure color count matches category count
)
plt.title('Cost Distribution by Shipping Carrier', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
carrier_revenue = df.groupby(['Shipping_carriers'])['Revenue_generated'].sum().reset_index()
carrier_revenue['Revenue_generated'] = carrier_revenue['Revenue_generated'].round(2)
carrier_revenue
```

```python
# Example: Group revenue by shipping carriers (create carrier_revenue if needed)
carrier_revenue = df.groupby('Shipping_carriers')[['Revenue_generated']].sum().reset_index()
carrier_revenue['Revenue_generated'] = carrier_revenue['Revenue_generated'].round(2)

# Plot
plt.figure(figsize=(10, 6))
p = sns.barplot(x='Shipping_carriers', y='Revenue_generated', data=carrier_revenue, palette='Wistia_r')

# Add value labels inside bars
for container in p.containers:
    p.bar_label(container, padding=-40, color='black', fontsize=10)

plt.title("Revenue Generated by Shipping Carriers")
plt.xlabel("Shipping Carriers")
plt.ylabel("Revenue Generated")
plt.tight_layout()
plt.show()
```

##### Both the graphs clearly show shipping carrier B is costly as well as generating higher revenue.

```python
transport = df.groupby(['Transportation_modes', 'Routes'])['Costs'].sum().reset_index()
transport
```

```python
# Group by transportation mode and route to summarize total costs
transport = df.groupby(['Transportation_modes', 'Routes'])[['Costs']].sum().reset_index()

# Round for clean display
transport['Costs'] = transport['Costs'].round(2)

# Plot
plt.figure(figsize=(12, 8))
p = sns.barplot(
    x='Transportation_modes',
    y='Costs',
    hue='Routes',
    data=transport,
    palette='twilight'
)

# Add bar labels (optional: comment out if grouped bars don't have containers)
for container in p.containers:
    p.bar_label(container, padding=-20, color='white', fontsize=10)

plt.title("Total Transportation Costs by Mode and Route")
plt.xlabel("Transportation Modes")
plt.ylabel("Total Costs")
plt.tight_layout()
plt.show()
```

```python
shipping = df.groupby(['Shipping_carriers', 'Transportation_modes'])['Shipping_times'].mean().reset_index()
shipping['Shipping_times'] = shipping['Shipping_times'].round(2)
shipping
```

```python
plt.figure(figsize = (12, 8))
p = sns.barplot(x = shipping['Transportation_modes'], y = shipping['Shipping_times'], hue = shipping['Shipping_carriers'], palette = 'viridis')
for container in p.containers:
    p.bar_label(container,padding=-20, color='white', fontsize=10)
plt.show()
```

According to the graph, the fastest and most efficient shipping option is Carrier B in all four transportation modes.
